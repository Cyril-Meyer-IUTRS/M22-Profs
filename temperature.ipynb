```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%cd data/temperature
```

```python
df = pd.read_csv('DECADAGRO_67_previous-1950-2022.csv.gz', sep=';')
```

```python
df = df[df.NOM_USUEL == 'STRASBOURG-ENTZHEIM']
```

```python
y_max = np.array(df['TX'])
y_min = np.array(df['TN'])
x = [i for i in range(len(y_max))]
```

```python
figure = plt.figure(figsize=(18, 4))
plt.plot(x, y_max, y_min)
```

```python
reg_max = np.polyfit(x, y_max, deg=1)
trend_max = np.polyval(reg_max, x)
reg_min = np.polyfit(x, y_min, deg=1)
trend_min = np.polyval(reg_min, x)

figure = plt.figure(figsize=(12, 4))
plt.scatter(x, y_max, color='#ff000010', marker='.')
plt.scatter(x, y_min, color='#0000ff10', marker='.')
plt.plot(x, trend_max, color='red')
plt.plot(x, trend_min, color='blue')
```

```python
print(round(trend_min[0], 2), round(trend_min[-1], 2), round(trend_min[-1] - trend_min[0], 2))
print(round(trend_max[0], 2), round(trend_max[-1], 2), round(trend_max[-1] - trend_max[0], 2))
```
